```python
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity','OnlineBackup', 'DeviceProtection', 
           'TechSupport', 'StreamingTV', 'StreamingMovies']
n_cols = 3
n_rows = 3
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20,20))  
axes = axes.flatten() 
for i, col in enumerate(columns):
    ax = axes[i]
    sns.countplot(x=col, data=df, hue="Churn", palette="pastel", width=0.6, ax=ax)
    ax.set_title(f"Churn by {col}")
    for bars in ax.containers:  
        ax.bar_label(bars)
plt.tight_layout()
plt.show()
```

**Findings about Impact of Services on Customer Churn: Analyzing Retention Factors**                               
**📊 Insights:**                                                                      
Each graph represents a service category (e.g., Phone Service, Internet Service, Streaming TV) and compares churn vs. non-churn customers.      
🔹 If more churned customers don’t have a service, it suggests the service is important for retention.                      
🔹 If more churned customers have the service, it may indicate dissatisfaction with it.

**🔎 Findings:**
**Phone Service & Multiple Lines:**                                                                                
If churned customers lack these, it suggests phone services are valuable.                            
If churn is high despite having them, it may indicate poor service quality.
       
**Internet Service:**                                                                            
If churn is higher among DSL or Fiber users, pricing or quality may be an issue.                                       
Customers without internet service may have fewer reasons to stay.

**Online Security, Backup, & Device Protection:**                                                                             
If customers with these services still churn, they may not see value in them.                                
If churned customers lack these, the company should promote their importance.

**Tech Support:**                                                                                           
Higher churn for those without tech support means support is a key retention factor.
Poor support may also drive customers away.

**Streaming TV & Movies**                                                                      
If subscribers churn, competitors may offer better entertainment packages.
If non-subscribers churn, bundling streaming services may increase retention.

**✅ Recommendations:**                                                                               
Highlight essential services like tech support & security to reduce churn.

Improve service quality where churn is high despite service usage.

Bundle services (e.g., Internet + Streaming) for better customer retention.

Offer better pricing or discounts for high-churn services.

**🛠 Methods to Improve:**                                                      
*Targeted promotions for customers without high-retention services.*

*Enhance service quality by gathering customer feedback.*

*AI-based customer support for faster resolutions.*

*Personalized plans based on user behavior.*
